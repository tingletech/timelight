```python
import itertools
import math
from IPython.core.display import display, HTML
import pylab
import numpy as np
```

```python
def colorize(minute, frequency=0.0041887902047863905, amplitude=127, center=128):
    value = math.sin((minute - 120) * frequency) * amplitude + center
    return int(round(value,1))

def adjust_color(minute):
    r = colorize(minute + 240)
    g = colorize(minute)
    b = colorize(minute - 240)
    x = colorize(minute - 480)
    return r, g, b, x
```

```python
%pylab inline

x = np.linspace(0, 60 * 24, 200)

y1, y2, y3, y4 = [], [], [], []
for i in x:
    r, g, b, X = adjust_color(i)
    y1.append(r)
    y2.append(g)
    y3.append(b)
    y4.append(X)


pylab.plot(x, np.array(y1), label='r')
pylab.plot(x, np.array(y2), label='g')
pylab.plot(x, np.array(y3), label='b')
pylab.plot(x, np.array(y4), label='x')
pylab.legend()
```

```python
out = ''
count = 0
for minute in range(0, 5000, 60):
    out = out + '<div style="background-color: rgb({0},{1},{2})">{0},{1},{2}|<b>{3}</b>|{4}|</div>'.format(*adjust_color(minute), minute)
    count = count + 1
display(HTML(out))
```
